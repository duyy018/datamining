# Data Mining
#### import library BeautifulSoup4 dan Request serta menyimpan alamat pada di variabel

```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
from selenium import webdriver
import re
import time
con2 = webdriver.Chrome()
html = "https://www.jd.id/search?keywords=laptop%2BAsus&sortType=sort_total_sale_amount_desc&lowPrice=2500000&highPrice=15000000"
```

##### Open Koneksi -> Download Page -> Close Koneksi -> Save to Variable -> Print beberapa isi Variable

```python
con = uReq(html)
html = con.read()
con. close
page = soup(html, "lxml")
print(type(page))
print(page.prettify()[1:200])
```

##### Mencari list Produk -> Print Panjang LIst Produk -> 

```python
p = page.findAll("div", {"data-suit":"0"})
print(len(p))
produk = p[0]
nama = produk.find("div",{"class":"p-desc"}).a["title"]
harga = produk.find("span",{"eptid":"price"})
diskon = produk.find("span",{"class":"discount-rate"})
pembeli = produk.find("div",{"class":"p-user-line clearfix"})
link = produk.find("div",{"class":"p-desc"}).a["href"].replace("//","https://")

print(link)
con2.get(link)
con2.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(2)
page2 = soup(con2.page_source, "html.parser")
print(page2.prettify()[1:15])
p2 = page2.findAll("div", {"class":"comment-item"})
print(len(p2))

comen = p2[8]
komen = comen.find("div",{"class":"item-content"}).text.strip()
orang = comen.find("div",{"class":"item-user"}).text.strip()
print(orang+" : "+komen)

kalimat = komen.lower().replace(".", " ").replace(","," ").split(" ")
cari = ['baik', 'ori']
for c in range(len(cari)):
    banyak = 0
    for i in range(len(kalimat)):
        if cari[c] in kalimat[i]:
            banyak = banyak +1
    print(cari[c]+" : "+str(banyak))
```

###### Membuat file Produk.csv -> Memberi Kolom

```python
namafile = "produk3.csv"
f = open(namafile, "w")
column="Nama_Produk,Warna,Harga,Diskon,Pembeli,orang,komentar,baik,ori\n"
f.write(column)
```

##### Looping cetak dan masukan data ke produk.csv

```python
for produk in p:
    
    Nama_Produk = produk.find("div",{"class":"p-desc"}).a["title"]
    Harga = produk.find("span",{"eptid":"price"})
    Diskon = produk.find("span",{"class":"discount-rate"})
    Pembeli = produk.find("div",{"class":"p-user-line clearfix"})
    link = produk.find("div",{"class":"p-desc"}).a["href"]
    
    #text processing
    pnama = Nama_Produk.split("/")[0].replace("\u200B","")
    pwarna = Nama_Produk.split("-")[-1].strip()
    pharga = Harga.text.split(" ")[1].replace(",","")
    ppembeli = Pembeli.text.strip().replace("(","").replace(")","")
    pdiskon = Diskon.text.split(" ")[0]
    plink = link.replace("//","https://")
    
    print(pnama+" | "+pwarna+" | Rp."+pharga+" | "+pdiskon+"% | "+ppembeli+" Orang"+"\n"+link)
    
    #open newtab
    con2.get(plink)
    con2.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    page2 = soup(con2.page_source, "html.parser")
    p2 = page2.findAll("div", {"class":"comment-item"})
    print(len(p2))
    for comen in p2:
        komen = comen.find("div",{"class":"item-content"}).text.strip()
        orang = comen.find("div",{"class":"item-user"}).text.strip()
        pkomen = re.sub('[^ a-zA-Z0-9]', '', komen)                                                                        
        print(orang+" : "+komen)
        
        f.write(pnama+","+pwarna+","+pharga+","+pdiskon+","+ppembeli+","+orang+","+pkomen)
        
        kalimat = komen.lower().replace("."," ").replace(","," ").split(" ")
        cari = ['baik', 'ori']
        for c in range(len(cari)):
            banyak = 0
            for i in range(len(kalimat)):
                if cari[c] in kalimat[i]:
                    banyak = banyak +1
            print(cari[c]+" : "+str(banyak))
            f.write(","+str(banyak))
        f.write("\n")
    
    print("\n")
f.close()
```
