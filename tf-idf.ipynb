```python
import csv
import math

#Cari File
with open('produk45.csv', newline='') as csvfile:
    data = list(csv.reader(csvfile))

#buat file csv bobot
namafile = "bobot.csv"
f = open(namafile, "w")
#header
column="df,D/df,log(D/df),log(D/df)+1"
f.write(column)
for h in range(len(data)-1):
    f.write(",w"+str(h+1))
f.write("\n")


    
#cari D   
d = len(data)
print("D : "+str(d-1))

#cari Df
for j in range(len(data[0])):
    df = 0
    if j > 2 :
        for i in range(len(data)):
            if data[i][j] != '0':
                df = df+1
        df=df-1
        print("Df ke-"+str(j)+" : "+str(df))
        
        #cari D/df
        if df != 0:
            ddf = d / df
        else:
            ddf = 0
        print("D/df : "+str(ddf))
        
        #cari log D/df
        logddf = round(math.log(ddf ,10), 2)
        print("log(D/Df) : "+str(logddf))
        
        #ldf + 1
        ldf1 = round(logddf + 1 , 2)
        print("log(D/Df) + 1  : "+str(ldf1))
        
        f.write(str(df)+","+str(ddf)+","+str(logddf)+","+str(ldf1))
        # W
        for k in range(len(data)):
            w = 0.0
            if k > 0 :
                w = int(data[k][j]) * ldf1
                print("W-d"+str(k)+" : "+str(w))
                f.write(","+str(w))
        print("\n")
        f.write("\n")
f.close()
```
